```python
import networkx as nx
import matplotlib.pyplot as plt

# Tạo một đồ thị mạng xã hội mới
G = nx.Graph()

# Thêm các nhân vật và mối quan hệ vào đồ thị
G.add_nodes_from([
    ("Harry Potter", {"house": "Gryffindor"}),
    ("Hermione Granger", {"house": "Gryffindor"}),
    ("Ron Weasley", {"house": "Gryffindor"}),
    ("Draco Malfoy", {"house": "Slytherin"}),
    ("Albus Dumbledore", {"house": "Gryffindor"})
])

G.add_edges_from([
    ("Harry Potter", "Hermione Granger"),
    ("Harry Potter", "Ron Weasley"),
    ("Harry Potter", "Draco Malfoy"),
    ("Hermione Granger", "Ron Weasley"),
    ("Hermione Granger", "Draco Malfoy"),
    ("Ron Weasley", "Draco Malfoy"),
    ("Harry Potter", "Albus Dumbledore"),
    ("Hermione Granger", "Albus Dumbledore"),
    ("Ron Weasley", "Albus Dumbledore")
])

# Vẽ đồ thị mạng xã hội
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True, node_size=1000, node_color='lightblue', edge_color='gray')

# Hiển thị thông tin về các nhân vật và mối quan hệ
plt.title("Harry Potter Social Network")
plt.show()
```
